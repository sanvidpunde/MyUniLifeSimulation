### Harmonic Mean
The Harmonic Mean (HM) is defined as the reciprocal of the average of the reciprocals of the data values.
It is based on all the observations, and it is rigidly defined.
Harmonic mean gives less weightage to the large values and large weightage to the small values to balance the values correctly

```python
print ("Geometric mean of OS marks: ",statistics.harmonic_mean(df['Acedamic percentage in Operating Systems']))
print ("Geometric mean of Algorithms marks: ",statistics.harmonic_mean(df['percentage in Algorithms']))
print ("Geometric mean of Programming Concepts marks: ",statistics.harmonic_mean(df['Percentage in Programming Concepts']))
print ("Geometric mean of Software Engineering marks: ",statistics.harmonic_mean(df['Percentage in Software Engineering']))
print ("Geometric mean of Computer Networks marks: ",statistics.harmonic_mean(df['Percentage in Computer Networks']))
print ("Geometric mean of Electronics marks: ",statistics.harmonic_mean(df['Percentage in Electronics Subjects']))
print ("Geometric mean of Comp Architecture marks: ",statistics.harmonic_mean(df['Percentage in Computer Architecture']))
print ("Geometric mean of Maths marks: ",statistics.harmonic_mean(df['Percentage in Mathematics']))
print ("Geometric mean of Communications marks: ",statistics.harmonic_mean(df['Percentage in Communication skills']))
print ("Geometric mean of Daily Working Hours: ",statistics.harmonic_mean(df['Hours working per day']))
print ("Geometric mean of Logical Quotient: ",statistics.harmonic_mean(df['Logical quotient rating']))
print ("Geometric mean of Hackathons: ",statistics.harmonic_mean(df['hackathons']))
print ("Geometric mean of Coding skills: ",statistics.harmonic_mean(df['coding skills rating']))
print ("Geometric mean of Public Speaking rating: ",statistics.harmonic_mean(df['public speaking points']))
```

### Confidence Interval

```python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
print(mean_confidence_interval(df['Acedamic percentage in Operating Systems'], confidence=0.95))
print(mean_confidence_interval(df['percentage in Algorithms'], confidence=0.95))
print(mean_confidence_interval(df['Percentage in Programming Concepts'], confidence=0.95))
print(mean_confidence_interval(df['Percentage in Software Engineering'], confidence=0.95))
print(mean_confidence_interval(df['Percentage in Computer Networks'], confidence=0.95))
print(mean_confidence_interval(df['Percentage in Electronics Subjects'], confidence=0.95))
print(mean_confidence_interval(df['Percentage in Computer Architecture'], confidence=0.95))
print(mean_confidence_interval(df['Percentage in Mathematics'], confidence=0.95))
print(mean_confidence_interval(df['Percentage in Communication skills'], confidence=0.95))
print(mean_confidence_interval(df['Hours working per day'], confidence=0.95))
print(mean_confidence_interval(df['Logical quotient rating'], confidence=0.95))
print(mean_confidence_interval(df['hackathons'], confidence=0.95))
print(mean_confidence_interval(df['coding skills rating'], confidence=0.95))
print(mean_confidence_interval(df['public speaking points'], confidence=0.95))
```

```python
for i in df['hackathons']:
    q1 = df['hackathons'].quantile(0.25)
    q3 = df['hackathons'].quantile(0.75)
    iqr = q3-q1
    lower_tail = q1 - 1.5*iqr
    upper_tail = q3 + 1.5*iqr
    
    if i>upper_tail or i<lower_tail:
        df['hackathons'] = df['hackathons'].replace(i,0)
        
sns.boxplot(df['hackathons'])
plt.title('Boxplot after removing Outliers')
plt.show()
```
