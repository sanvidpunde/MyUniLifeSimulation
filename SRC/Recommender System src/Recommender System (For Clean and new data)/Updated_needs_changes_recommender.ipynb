```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.naive_bayes import GaussianNB
import warnings
warnings.filterwarnings("ignore")
```

```python
def CAO_SCORE(score):
    if  (score >= 50)  and (score <250 ) :
        return "Low"
    elif  (score >= 250)  and (score <500 ) :
        return "Medium"
    elif (score >= 500):
        return "High"
    #else: return 'Neutral'

df['cao_score'] = df['cao_score'].apply(CAO_SCORE)
```

```python
df ['fees'] = df ['fees'].astype(float)
```

```python
#df ['fees']=df.fees.fillna(df.fees.mean(), inplace=True)
```

```python
def Budget(Budget):
    if  (Budget > 2000)  and (Budget <4000 ) :
        return "Low"
    elif  (Budget > 4001)  and (Budget <6000 ) :
        return "Medium"
    elif (Budget > 6001)  and (Budget <8000 ) :
        return "High"
    elif (Budget > 8001) :
        return "Very High"
    else: return 'High'

df['budget'] = df['fees'].apply(Budget)
```

```python
feed = df[['cao_score', 'budget', 'location_code', 'interest_code', 'job_domain_code','code']]

# Choosing independent vars
df_train_x = feed.drop('code',axis = 1)

# Choosing all target vars
df_train_y = feed['code']

x_train, x_test, y_train, y_test = train_test_split(df_train_x, df_train_y, test_size=.90,random_state=42)
```

```python
y_train = pd.DataFrame(y_train, columns = ["code"])
y_train.head()
y_test = pd.DataFrame(y_test, columns = ["code"])
y_test
```

```python
#Decison Tree
dtree = DecisionTreeClassifier(random_state=60)
dtree = dtree.fit(x_train, y_train)
y_pred = dtree.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test,y_pred)
print("confusion matrics=",cm)
print("  ")
print("accuracy=",accuracy)
```

```python

userdata = [[1,0,4,4,1]]
ynewclass = dtree.predict(userdata)
ynew = dtree.predict_proba(userdata)
print(ynewclass)
print("Probabilities of all classes: ", ynew)
print("Probability of Predicted class : ", np.max(ynew))
```

```python
##Random Forest
rf = RandomForestClassifier(random_state = 50)
rf.fit(x_train, y_train)
rfc_y_pred = rf.predict(x_test)
rfc_cm = confusion_matrix(y_test,rfc_y_pred)
rfc_accuracy = accuracy_score(y_test,rfc_y_pred)
print("confusion matrics=",rfc_cm)
print("  ")
print("accuracy=",rfc_accuracy)
```
